```python
import itertools
import matplotlib.patches as mpl_patches
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import spearmanr, kendalltau, pearsonr


from sklearn.datasets import  load_breast_cancer
```

```python
def correlaciones(dataset,columns):
    """
    OBJETIVO Y OUTPUT DE LA FUNCIÓN: 
    * Mostrar las correlaciones entre columnas numéricas continuas deseadas de un dataframe.
    * Cada fila representa la relación entre dos variables y contiene tres gráficos donde:
       - Izquierda: Gráfico de dispersión entre las mismas.
       - Centro y derecha: Distribución de las variables continuas utilizando histogramas.
    * Adicionalmente el gráfico de dispersión muestra un recuadro de texto con 
    el resultado de los coeficientes de correlación de Pearson, Spearman y Kendall.
    
    ENTRADAS:
       * dataset 
           - Dataframe con el cual se desea trabajar.
       * columns 
           - Lista de columnas que se desean analizar

       """
    #Se generan las combinaciones entre variables
    combinations = list(itertools.combinations(columns, 2)) 

    figura,axes = plt.subplots(len(combinations),3, figsize=(20,(20*columns.nunique()/3)))
    
    for i, j in zip(combinations, range(len(combinations))):
        sns.scatterplot(dataset[i[0]],dataset[i[1]], ax = axes[j,0])
        axes[j,0].grid()
        axes[j,0].tick_params(labelleft=True)
        sns.distplot(dataset[i[0]], ax=axes[j,1]).set_title(i[0])
        axes[j,1].grid()
        axes[j,1].set_xlabel('') 
        sns.distplot(dataset[i[1]], ax=axes[j,2]).set_title(i[1])
        axes[j,2].grid()
        axes[j,2].set_xlabel('')
        #Para mostrar los resultados de lo coeficientes de correlación
        handles = [mpl_patches.Rectangle((0, 0), 1, 1, fc="white", ec="white", 
                                 lw=0, alpha=0)] * 3
        labels = []
        labels.append("pearson = {0:.4g}".format(pearsonr(dataset[i[0]], 
                                                    dataset[i[1]])[0]))
        labels.append("spearman = {0:.4g}".format(spearmanr(dataset[i[0]], 
                                                    dataset[i[1]])[0]))
        labels.append("kendall = {0:.4g}".format(kendalltau(dataset[i[0]], 
                                                    dataset[i[1]])[0]))
        axes[j,0].legend(handles, labels, loc='upper right', fontsize='small', 
          fancybox=True, framealpha=0.7, 
          handlelength=0, handletextpad=0)
    figura.subplots_adjust(hspace=.5) 
    plt.show()
```

Para visualizar como trabaja la función utilizaremos el dataset de cancer de sklearn.

```python
cancer = load_breast_cancer()
cancer_df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
cancer_df['target'] = pd.Series(cancer.target)
cancer_df.head()
```

Como el dataset tiene muchas columnas, a modo de ejemplo solo utilizaremos las 6 primeras.

```python
cancer_df.columns[:5]
```

```python
correlaciones(cancer_df,cancer_df.columns[:5])
```
