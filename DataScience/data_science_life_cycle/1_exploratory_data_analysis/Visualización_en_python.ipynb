## Figuras con matplotlib

En actualidad, mayormente la visualizaciones de matplotlib se realizan utilizando el [módulo pyplot](https://matplotlib.org/3.1.1/tutorials/introductory/pyplot.html), que es una colección de funciones de estilos de comando basada en los comandos de matlab. Esa es la razón por la que solemos realizar la importación:

<code>import matplotlib.pyplot as plt</code>

y trabajar directamente con plt.

Por otro lado, un concepto importante al momento de utilizar matplotlib es su jerarquía de objetos, esto significa que cada objeto dibujado es una estructura que a su vez encapsula a otra como muestran las imágenes a continuación (las mismas fueron extraídas de [link1](https://matplotlib.org/) y [link2](https://realpython.com/python-matplotlib-guide/)).

Así, el objeto **Figure** es lo que encapsula todo y a su vez cada figura puede contener múltiples objetos **Axes** que serían como gráficos individuales dentro de una misma figura. Un objecto Axes por ejemplo encapsula a su vez los ejes (x e y) y sus etiquetas, leyendas, cajas con texto, marcas, etc.

### Creación de figuras: Configuración inicial

Al momento de generar una figura por primera vez existen ciertos comandos básicos y no tan basicos como:

- <code>plt.figure()</code>: Creación de la figura
- <code>plt.title()</code>: Adición del título
- <code>plt.legend()</code>: Adición de una leyenda
- <code>plt.grid()</code>: Generación de una grilla de líneas.
- <code>plt.xlabel()</code>: Adición de etiqueta en el eje x.
- <code>plt.ylabel()</code>: Adición de etiqueta en el eje y.
- <code>plt.xticks()</code>: Obtiene o establece las ubicaciones y etiquetas en el eje x.
- <code>plt.yticks()</code>: Obtiene o establece las ubicaciones y etiquetas en el eje y.

o sus análogas cuando trabajamos con axes:

- <code>axes.set_title()</code>
- <code>axes.set_label()</code>
- <code>axes.grid()</code>
- <code>axes.xaxis.set_tick_params()</code> ó <code>axes.yaxis.set_tick_params()</code>

Veamos un ejemplo:

```python
plt.figure()
#Titulo
plt.title('TITULO')
#Nombres de los ejes
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
#Rotacion en los valores de los ejes
plt.xticks(rotation=45)
plt.yticks(rotation=45)
#Leyenda (En este caso no tenemos ninguna declarada)
plt.legend()
plt.grid()
```

Observemos que el warning "*No handles with labels found to put in legend*" hace referencia a que no tenemos ninguna leyenda para colocar en el gráfico.

#### Otros comandos útiles

- <code>fig.subplots_adjust(hspace=.5)</code> ó bien  <code>fig.subplots_adjust(vspace=.5)</code>: Genera un espaciamiento  horizontal o vertical respectivamente entre los gráficos.
- <code>plt.tight_layout()</code>: Relleno entre el borde la figura y los bordes de los subplots.
- <code>plt.suptitle()</code>: Agrega un título centrado a la figura principal.
- <code>axes.set_xlim(a,b)</code> y <code>axes.set_ylim(c,d)</code>: Configura los límites superiores e inferiores en los valores que pueden tomar los ejes x e y respectivamente.
- <code>axes.yaxis.tick_right()</code>: Cambia la posición del eje vertical a la derecha.



### Creación de subfiguras

Muchas veces nos encontramos con la necesidad de realizar comparaciones entre diferentes gráficos y tener que hacer uno por figura puede resultar bastante tedioso o ineficiente.

Para ello matplotlib dispone de diferentes funciones para realizar subfiguras dentro de una misma figura como veremos a continuación.

#### <code>plt.subplot()</code>

[<code>plt.subplot()</code>](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.subplot.html#matplotlib.pyplot.subplot) añade subplots a una determinada figura. 

La sintaxis básica es 

<center><code>plt.subplot(a,b,c)</code> ó bien <code>plt.subplot(abc)</code> </center> 

donde a,b y c son números enteros y además:

- a: número de filas
- b: número de columnas
- c: índice (que comienza en la esquina superior izquierda y se mueve hacia la derecha)

**VENTAJA**
- Rápido si se quieren realizar pocos subplots.
- La grilla y los subplots se definen de antemano (¿desventaja?).

**DESVENTAJA**
- Los subplots se añaden uno a uno lo cual puede hacerse bastante tedioso si tenemos muchos subgráficos.
- Puede resultar complicado si queremos implementarlo dentro de un for (nunca lo intenté!)

```python
#EJEMPLO 1
fig = plt.figure(1)

ax1 = plt.subplot(231)
ax2 = plt.subplot(232)
ax3 = plt.subplot(233)
ax4 = plt.subplot(234)
ax5 = plt.subplot(235)
ax6 = plt.subplot(236)

#Titulos de cada uno de los graficos
ax1.set_title('Gráfico 1')
ax2.set_title('Gráfico 2')
ax3.set_title('Gráfico 3')
ax4.set_title('Gráfico 4')
ax5.set_title('Gráfico 5')
ax6.set_title('Gráfico 6')

plt.show()
```

```python
#EJEMPLO 2
fig = plt.figure(1)

ax1 = plt.subplot(211)
ax1.set_title('Gráfico 1')

ax2 = plt.subplot(223)
ax2.set_title('Gráfico 2')

plt.show()
```

#### <code>plt.subplots()</code>

[<code>plt.subplots()</code>](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots.htmll) crea una figura y una grilla de subfiguras con una simple línea de código. Permite un buen control de los gráficos individuales. Devuelve una figura y un objeto Axes o un arreglo de objectos Axes. 

La sintaxis básica es 

<center><code>plt.subplots(nrows=a,ncols=b)</code> ó bien <code>plt.subplots(a,b)</code> </center> 

donde a,b y c son números enteros y además:

- a: número de filas
- b: número de columnas



**VENTAJA**
- Rápido manejo de la grilla de subfiguras.
- Se podría utilizar adentro de un for para generar los subplots.

**DESVENTAJA**
- Todas las subfiguras tienen el mismo tamaño.

En el caso de gráficos y subgráficos en una sola dirección (todo horizontal o todo vertical por ejemplo) <code>axes</code> es un arreglo de numpy 1D.

[Link con información útil](https://www.python-course.eu/matplotlib_subplots.php).


```python
#EJEMPLO 1
fig, axes = plt.subplots(2)

#Título principal
fig.suptitle('TÍTULO')

#Titulos en los graficos
axes[0].set_title('Gráfico 1')
axes[1].set_title('Gráfico 2')

#Grillas
axes[0].grid()
axes[1].grid()

fig.subplots_adjust(hspace=.5) #Generamos un espaciamiento  horizontal entre los gráficos
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))
fig.subplots_adjust(wspace=.2) 


ax1.set_title('Título gráfico 1')
ax1.set_xlabel('Variable 1.1')
ax1.set_ylabel('Variable 2.1')
ax1.set_xlim(1,10)
ax1.set_ylim(2,8)

ax2.set_title('Título gráfico 2')
ax2.set_xlabel('Variable 1.2')
ax2.set_ylabel('Variable 2.2')
ax2.yaxis.tick_right() # Mueve el eje vertical a la derecha
```

Cuando se plotea en dos direcciones <code>axes</code> es un arreglo de numpy en 2D.

####  <code>plt.subplot2grid()</code>

[<code>plt.subplot2grid()</code>](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.subplot2grid.html) crea un eje (axis) y una locación específica dentro de un grilla regular. 

La sintaxis básica es 

<center><code>plt.subplot2grid(shape,loc)</code> </center> 

donde 

- shape: es la dimensión de la grilla
- loc: la posición en la grilla

**VENTAJAS:**
- Podemos añadir grillas de diferentes tamaños.
- Se podría utilizar adentro de un for para generar los subplots.
- Es posible que una figura ocupe 2 o más lugares.

**DESVENTAJAS:**
- Puede resultar tedioso pensar en las ubicaciones dentro de la grilla.

Veamos un ejemplo simple (que tranquilamente podríamos hacerlo también con <code>plt.subplots()</code>):

```python
#EJEMPLO 1
fig = plt.figure(figsize=(20,5))
ax1 = plt.subplot2grid((1,3),(0,0))
#Gráfico 1
plt.title('TITULO GRÁFICO 1')
ax2 = plt.subplot2grid((1,3),(0,1))
#Gráfico 2
plt.title('TITULO GRÁFICO 2')
ax3 = plt.subplot2grid((1,3),(0,2))
#Gráfico 3
fig.suptitle('TÍTULO GENERAL') 
plt.title('TITULO GRÁFICO 3')
plt.show()
```

Veamos ahora las ventajas de usar subplot2grid para crear subplots de diferentes dimensiones.

```python
#EJEMPLO 2
fig = plt.figure(figsize=(20,5))
#Gráfico 1: Posición (0,0) creo una grilla 3x3
ax1 = plt.subplot2grid((3,3),(0,0))
plt.title('TITULO GRÁFICO 1')
#Gráfico 2: Posición (0,1) creo una grilla 1x3
ax2 = plt.subplot2grid((1,3),(0,1))
plt.title('TITULO GRÁFICO 2')
#Gráfico 3: Posición (0,2) creo una grilla 3x3
plt.subplot2grid((3,3),(0,2))
ax3 = fig.suptitle('TÍTULO GENERAL') 
plt.title('TITULO GRÁFICO 3')
plt.show()
```

y quizás lo más interesante...

#### <code>plt.gridspec()</code>

[<code>plt.gridspec()</code>](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.gridspec.GridSpec.html) es una alternativa a los subplots que especifica de antemano la geometría de los subplots a ser creados. De esta manera, se crea una grilla con un número de filas y columnas los cuales se van rellenando a medida que se necesita.

##### Combinado con <code>plt.subplot()</code>

La siguiente imagen está inspirada en un Ejemplo extraído de [aquí](https://www.python-course.eu/matplotlib_subplots.php) en la cual crearemos diferentes subplots dentro de una grilla 3x3.

```python
plt.figure(figsize=(6, 4))
G = plt.GridSpec(3, 3)

ax1 = plt.subplot(G[0, :])
ax1.set_xticks(())
ax1.set_yticks(())
ax1.text(0.5, 0.5, 'Axes 1', ha='center', va='center', size=24, alpha=.5)
ax1.set_title('Grilla[0,:]')

ax2 = plt.subplot(G[1, :-1])
ax2.set_xticks(())
ax2.set_yticks(())
ax2.text(0.5, 0.5, 'Axes 2', ha='center', va='center', size=24, alpha=.5)
ax2.set_title('Grilla[1,:-1]')

ax3 = plt.subplot(G[1:, -1])
ax3.set_xticks(())
ax3.set_yticks(())
ax3.text(0.5, 0.5, 'Axes 3', ha='center', va='center', size=24, alpha=.5)
ax3.set_title('Grilla[1:,-1]')

ax4 = plt.subplot(G[-1, 0])
ax4.set_xticks(())
ax4.set_yticks(())
ax4.text(0.5, 0.5, 'Axes 4', ha='center', va='center', size=24, alpha=.5)
ax4.set_title('Grilla[-1,0]')


ax5 = plt.subplot(G[-1, -2])
ax5.set_xticks(())
ax5.set_yticks(())
ax5.text(0.5, 0.5, 'Axes 5', ha='center', va='center', size=24, alpha=.5)
ax5.set_title('Grilla[-1,-2]')

plt.tight_layout()
plt.suptitle('Grilla[3,3]',y=1.05,fontsize=20)
plt.show()
```

##### Combinado con <code>plt.subplots()</code> y <code>plt.subplot()</code>

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=1, ncols=4,figsize=(11, 7))

grid = plt.GridSpec(4, 2, wspace=0.2, hspace=0.5)

ax1 = plt.subplot(grid[0, :])
ax2 = plt.subplot(grid[1:3, 0])
ax3 = plt.subplot(grid[1:3,-1])
ax4 = plt.subplot(grid[3, :])

ax1.title.set_text('First Plot')
ax2.title.set_text('Second Plot')
ax3.title.set_text('Third Plot')
ax4.title.set_text('Fourth Plot')

plt.show()
```

Finalmente algo interesante es realizar una figura dentro de otra. En este caso se crea una única figura con dos <code>Axes</code> asociados.

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.9, 0.9]) # eje principal
axes2 = fig.add_axes([0.2, 0.6, 0.4, 0.3]) # eje secundario

# Figura principal
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Título figura principal')

# Figura secundaria
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('Título subfigura');
```
